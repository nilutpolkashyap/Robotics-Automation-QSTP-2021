## Linear Activation Function

```python
def linear_activation_function(a, x):
    return a*x
```

```python
x = np.linspace(-10, 10, 5000)
a1 = 1
y1 = []
a2 = 2
y2 = []
for i in x:
    y1.append(linear_activation_function(a1,i))
    y2.append(linear_activation_function(a2,i))
print(y1)
print(y2)
```

```python
plt.plot(x,y1,'g')
plt.plot(x,y2,'r')
plt.title('linear activation function')
plt.legend(('a = 1', 'a = 2'))
plt.show()
```

## Sigmoid Activation Function

```python
def sigmoid_activation_function(x):
    res = 1.0/(1 + np.exp(-x))
    return res
```

```python
x = np.linspace(-10, 10, 5000)
y = []
for i in x:
    y.append(sigmoid_activation_function(i))
print(y)
```

```python
plt.plot(x,y)
plt.title('sigmoid activation function')
plt.show()
```

## tanh activation function

```python
def tanh_activation_function(x):
    res = 2*sigmoid_activation_function(2*x) -1
    return res
```

```python
x = np.linspace(-10, 10, 5000)
y = []
for i in x:
    y.append(tanh_activation_function(i))
print(y)
```

```python
plt.plot(x, y)
plt.title('tanh activation function')
plt.grid(True)
plt.show()
```

## ReLU activation Function

```python
def relu_activation_function(x):
    res = np.array([0,x]).max()
    return res
```

```python
x = np.linspace(-10,10,5000)
y = []
for i in x:
    y.append(relu_activation_function(i))
print(y)
```

```python
plt.plot(x,y)
plt.grid(True)
plt.title('relu activation function')
plt.show()
```

## Leaky ReLU Activation Function

```python
def leaky_relu_activation_function(x):
    if x< 0:
        res = 0.01*x
    else:
        res = x
    return res
```

```python
x = np.linspace(-10,10,5000)
y = []
for i in x:
    y.append(leaky_relu_activation_function(i))
print(y)
```

```python
plt.plot(x,y)
plt.grid(True)
plt.title('Leaky relu activation function')
plt.show()
```

## Parametric ReLU Activation Function

```python
def parametic_relu_activation_function(a,x):
    if x < 0.0:
        res = a*x
    else:
        res = x
    return res
```

```python
x = np.linspace(-10,10,5000)
a1, a2, a3, a4 = 0.25, 0.5, 0.75, 1.00
y1 = []
y2 = []
y3 = []
y4 = []
for i in x:
    y1.append(parametic_relu_activation_function(a1,i))
    y2.append(parametic_relu_activation_function(a2,i))
    y3.append(parametic_relu_activation_function(a3,i))
    y4.append(parametic_relu_activation_function(a4,i))
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.grid(True)
plt.title('parametric relu activation function')
plt.legend(('a = 0.25', 'a = 0.5', 'a = 0.75', 'a = 1.00'))
plt.show()
```

## Exponential eLU Activation Function

```python
def elu_activation_function(a,x):
    if x < 0:
        return a*np.exp(x) - 1
    else:
        return x
```

```python
x = np.linspace(-10,10,5000)
a1, a2, a3, a4 = 0.1, 1, 2, 5
y1 = []
y2 = []
y3 = []
y4 = []
for i in x:
    y1.append(elu_activation_function(a1,i))
    y2.append(elu_activation_function(a2,i))
    y3.append(elu_activation_function(a3,i))
    y4.append(elu_activation_function(a4,i))
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.grid(True)
plt.title('exponential linear unit')
plt.legend(('a = 0.1', 'a = 1', 'a = 2', 'a = 5'))
plt.show()
```
